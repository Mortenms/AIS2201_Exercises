```python
import matplotlib.pyplot as plt
import numpy as np

w = np.linspace(0, np.pi, 512)
z = np.exp(1j*w)
Hw = (0.5+0.5*z**-1)/(1-0.64*z**-2)

Hw_mag = np.abs(Hw)
Hw_phase = np.unwrap(np.angle(Hw))

plt.close(1);plt.figure(1,figsize=(10,5))
plt.subplot(2,1,1)
plt.plot(w/np.pi,Hw_mag)
plt.subplot(2,1,2)
plt.plot(w/np.pi, Hw_phase)
```
