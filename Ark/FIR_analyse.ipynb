```python
import matplotlib.pyplot as plt
import numpy as np

M = 37

n = np.arange(M)

wc = 0.3*np.pi

hn = wc/np.pi*np.sinc(wc/np.pi*(n-(M-1/2)))

plt.stem(hn)

```
