```python
import matplotlib.pyplot as plt
import numpy as np

N = 128

Xm = np.zeros(N) + 0j

freq_rad = np.pi/32
m_freq = freq_rad/(2*np.pi)*N
m_freq = int(m_freq)

Xm[m_freq] = -1j*N/2
Xm[N-m_freq] = 1j*N/2

xn = np.fft.ifft(Xm)
#sum(abs(np.imag(xn)))
xn = np.real(xn)

plt.close(2); plt.figure(2)
plt.stem(xn)
```

```python
##TIDSFORSKYVING

Xm = np.fft.fft(xn)
m = np.arange(len(Xm))

Ym = Xm*np.exp(2j*np.pi*-12*m/N)

yn = np.fft.ifft(Ym)

plt.close(2); plt.figure(2)
plt.stem(xn)
plt.stem(yn)
```

```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt

N = 64
n = np.arange(0, N)

# Generating a square wave with the correct frequency
x_n = sig.square(2 * np.pi * n / 32)

# Compute FFT of the signal
X_m = np.fft.rfft(x_n)
m = np.arange(len(X_m))

# Define cutoff for low-pass and high-pass filters
m_cutoff = 32

# Create the low-pass and high-pass filter masks
H_lp = (m < m_cutoff)
H_hp = (m > m_cutoff)

# Apply the low-pass filter
X_lp = X_m * H_lp
y_n = np.fft.irfft(X_lp)

# Plotting the original and filtered signals
plt.close(5)
plt.figure(5)

plt.subplot(2, 1, 1)
plt.stem(n, x_n, use_line_collection=True)
plt.title('Original Signal')

plt.subplot(2, 1, 2)
plt.stem(n[:len(y_n)], y_n, use_line_collection=True)
plt.title('Filtered Signal (Low-pass)')

plt.tight_layout()
plt.show()
```
