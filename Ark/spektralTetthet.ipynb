```python
import matplotlib.pyplot as plt
from scipy.io import wavfile
import scipy.signal as sig
import numpy as np
from IPython.display import Audio

fs, data = wavfile.read(r'piano.wav')
xn = data/0x8000
N = len(xn)

Xm = np.fft.rfft(xn)
P_xx = 1/(N*fs)*np.abs(Xm)**2
P_xx[1:-1] *=2
f_m = np.arange(0,len(P_xx))*fs/N

plt.close(1); plt.figure(1)
plt.plot(f_m,10*np.log10(P_xx))

f_m, P_xx2 = sig.periodogram(xn,fs=fs)
plt.plot(f_m,10*np.log10(P_xx2), ':')
plt.xlim((0,5000))
plt.ylim((-100,0))
```
